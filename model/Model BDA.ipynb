```python
import pickle
df = pickle.load(open('stock_data.pkl', 'rb'))
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import *
from keras.layers import *
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from keras.utils import plot_model
```

```python
def prepare_data(ticker, stock):
    
    data = stock.filter([ticker]).values
    scaler = MinMaxScaler(feature_range=(0,1))
    scaled_data = scaler.fit_transform(data)
    training_data_len = int(len(data) * 0.8)
    train_data = scaled_data[0:training_data_len, :]
    x_train = []
    y_train = []
    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i, 0])
        y_train.append(train_data[i, 0])
    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    test_data = scaled_data[training_data_len-60:, :]
    x_test = []
    y_test = data[training_data_len:, :]
    for i in range(60, len(test_data)):
        x_test.append(test_data[i-60:i, 0])
    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
    return x_train, x_test, y_train, y_test, scaler
```

```python
scalers = {}
X_trains = {}
X_tests = {}
y_trains = {}
y_tests = {}
for ticker in df.columns.values:
    X_trains[ticker], X_tests[ticker], y_trains[ticker], y_tests[ticker], scalers[ticker] = prepare_data(ticker, df)
```

```python
inp_shape = X_trains['AAPL'].shape[1]
inp_shape
```

```python
input = Input(shape=(inp_shape, 1), name='input')
lstm1 = LSTM(64, return_sequences=True, name='lstm_1')(input)
drop1 = Dropout(0.3, name='dropout_1')(lstm1)
lstm2 = LSTM(32, return_sequences=True, name='lstm_2')(drop1)
drop2 = Dropout(0.3, name='dropout_2')(lstm2)
outs = []
for key in df.columns.values:
    lstm = LSTM(16, return_sequences=False, name=f'lstm_{key}')(drop2)
    dense = Dense(64, name=f'dense_{key}')(lstm)
    out = Dense(1, name=f'output_{key}')(dense)
    out.trainable = False
    lstm.trainable = False
    dense.trainable = False
    outs.append(out)
```

```python
model = tf.keras.Model(inputs=input, outputs = outs)
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
plot_model(model, show_shapes=True, show_layer_names=True, dpi=400)
```

```python
model.summary()
```

```python
n = len(df.columns)
for i, key in enumerate(df.columns):
    print(f"TRAINING ON {key}, Output Layer {i+1}/{n}")
    X_train = X_trains[key]
    y_train = y_trains[key]
    X_test = X_tests[key]
    y_test = y_tests[key]
    scaler = scalers[key]
    model.get_layer(f"lstm_{key}").trainable = True
    model.get_layer(f"dense_{key}").trainable = True
    model.get_layer(f"output_{key}").trainable = True
    model.fit(X_train, y_train, epochs=20, batch_size=64, shuffle=False)
    model.get_layer(f"lstm_{key}").trainable = False
    model.get_layer(f"dense_{key}").trainable = False
    model.get_layer(f"output_{key}").trainable = False
    print("=========================================================================================\n")
```

```python
model.save("prediction_model.h5")
```
